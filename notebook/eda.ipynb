# Section 2: Univariate Analysis – Numerical, Categorical

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(df_cleaned['monthly_salary'], bins=50, kde=True)
plt.title('Monthly Salary Distribution')

plt.subplot(1, 2, 2)
sns.boxplot(x=df_cleaned['monthly_salary'])
plt.title('Monthly Salary Boxplot')
plt.show()
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(df_cleaned['age'], bins=50, kde=True)
plt.title('Age Distribution')

plt.subplot(1, 2, 2)
sns.boxplot(x=df_cleaned['age'])
plt.title('Age Boxplot')
plt.show()
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(df_cleaned['years_of_employment'], bins=50, kde=True)
plt.title('Years of employment Distribution')

plt.subplot(1, 2, 2)
sns.boxplot(x=df_cleaned['years_of_employment'])
plt.title('Years of employment Boxplot')
plt.show()
```

- Age: Highly discrete; most records are clustered around a few integer values (e.g., 32, 38, 48, etc.), indicating perhaps policy or demographic grouping by age.​

- Monthly Salary & Earnings: Right-skewed distribution with visible outliers in the upper tail. Most salaries cluster below ₹75,000, with a steep drop-off at higher values. Presence of substantial outliers above ₹2,00,000 requires attention in feature engineering.​

- Years of Employment: Most applicants have less than 10 years, with a concentration around zero (new employees or self-employed) and a long tail of higher experience. Strong right skew and several outliers above 20 years support using robust statistics over mean.​

- Credit Score: Boxplots by EMI eligibility demonstrate higher median scores for “Eligible”, lower for “Not_Eligible”, and more outliers for both extremes. This supports credit score as a strong risk differentiator.

- Gender: There is a significant gender imbalance (more males than females) in the data, which could indicate differences in loan applications or approval rates between genders.​

- EMI Eligibility: "Not_Eligible" makes up the majority of cases, followed by "Eligible", with "High_Risk" applicants as the smallest group, confirming a heavily imbalanced classification target distribution.​

# Section 6: Correlation Analysis & Feature Selection

```python
plt.figure(figsize=(16,12))
sns.heatmap(df_cleaned[num_cols].corr(), annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Feature Correlation Matrix')
plt.show()
```

# Correlation Matrix (Relationships)
- Most financial features show moderate positive or negative correlations.

- Highest pairwise correlations: groceries_utilities and school_fees, college_fees and travel_expenses, suggesting family size or life stage clusters.

- Max EMI correlates positively with monthly_salary, requested_amount, and bank_balance, negatively with current_emi_amount.​

- Credit Score has negative correlation with current EMI and school/college fees, suggesting credit stress indicators.

# Section 7: Outlier Summary

```python
for col in num_cols:
    q1 = df_cleaned[col].quantile(0.25)
    q3 = df_cleaned[col].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    outlier_count = len(df_cleaned[(df_cleaned[col] < lower) | (df_cleaned[col] > upper)])
    print(f"{col}: {outlier_count} outliers")
```
